```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Assuming X_train, X_test, Y_train, Y_test are defined elsewhere in your code
# Standardize features
std = StandardScaler()
X_train_std = std.fit_transform(X_train)
X_test_std = std.transform(X_test)

# Initialize KNN classifier
knn = KNeighborsClassifier()

# Train the classifier
knn.fit(X_train_std, Y_train)

# Predict on the test set
Y_pred = knn.predict(X_test_std)

# Calculate accuracy
knn_ac = accuracy_score(Y_test, Y_pred)
knn_ac
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Create a Random Forest Regression model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the Random Forest Regression model
rf_regressor.fit(X_train_std, Y_train)

# Predict on the test set
Y_pred_rf = rf_regressor.predict(X_test_std)

# Evaluate the Random Forest Regression model
mse_rf = mean_squared_error(Y_test, Y_pred_rf)
print("Mean Squared Error (Random Forest):", mse_rf)
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Mean Absolute Error (MAE)
mae_rf = mean_absolute_error(Y_test, Y_pred_rf)
print("Mean Absolute Error (Random Forest):", mae_rf)

# Root Mean Squared Error (RMSE)
rmse_rf = mean_squared_error(Y_test, Y_pred_rf, squared=False)
print("Root Mean Squared Error (Random Forest):", rmse_rf)

# R-squared (R2) score
r2_rf = r2_score(Y_test, Y_pred_rf)
print("R-squared (Random Forest):", r2_rf)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score

# Assuming X_train_std, X_test_std, Y_train are defined elsewhere in your code

# Convert data to DMatrix format (optimized for XGBoost)
dtrain = xgb.DMatrix(X_train_std, label=Y_train)
dtest = xgb.DMatrix(X_test_std)

# Set XGBoost parameters
param = {
    'max_depth': 6,  # Maximum depth of the tree
    'eta': 0.3,  # Learning rate
    'objective': 'multi:softmax',  # Objective function for multi-class classification
    'num_class': len(set(Y_train))  # Number of classes
}

# Train the model
num_round = 100  # Number of boosting rounds
xgb_model = xgb.train(param, dtrain, num_round)

# Make predictions
Y_pred_xgb = xgb_model.predict(dtest)

# Convert predictions to integers
Y_pred_xgb = Y_pred_xgb.astype(int)

# Calculate accuracy
xgb_ac = accuracy_score(Y_test, Y_pred_xgb)
print("Accuracy of XGBoost Classifier:", xgb_ac)
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score

# Assuming X_train_std, X_test_std, Y_train are defined elsewhere in your code

# Convert data to lightgbm Dataset format
train_data = lgb.Dataset(X_train_std, label=Y_train)
test_data = lgb.Dataset(X_test_std, label=Y_test)

# Set parameters for LightGBM
params = {
    'objective': 'multiclass',  # Specify the task as multiclass classification
    'num_class': len(set(Y_train)),  # Number of classes in the target variable
    'metric': 'multi_logloss',  # Evaluation metric
    'verbosity': -1,  # Set verbosity to -1 to suppress messages
}

# Train the model
num_round = 100  # Number of boosting rounds
lgb_model = lgb.train(params, train_data, num_round)

# Make predictions on the test set
Y_pred_lgb = lgb_model.predict(X_test_std)

# Convert predicted probabilities to class labels
Y_pred_lgb_class = [np.argmax(pred) for pred in Y_pred_lgb]

# Calculate accuracy
lgb_ac = accuracy_score(Y_test, Y_pred_lgb_class)
print("Accuracy of LightGBM Classifier:", lgb_ac)
```

```python
import matplotlib.pyplot as plt

# Define the classifiers and their corresponding accuracy scores
classifiers = ['Decision Tree', 'KNN', 'Logistic Regression', 'Random Forest', 'SVM', 'GBM', 'NN', 'Naive Bayes', 'Gaussian Process', 'XGBoost', 'LightGBM', 'Extra Trees']
accuracies = [dt_ac, knn_ac, lr_ac, rf_ac, svm_ac, gbm_ac, nn_ac, nb_ac, gpc_ac, xgb_ac, lgb_ac, extra_trees_ac]

# Define colors for each classifier
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'lime']

# Plot the accuracy scores with different colors
plt.figure(figsize=(12, 8))
bars = plt.barh(classifiers, accuracies, color=colors)
plt.xlabel('Accuracy')
plt.title('Accuracy of Different Classifiers')
plt.xlim(0, 1)  # Set x-axis limit from 0 to 1
plt.gca().invert_yaxis()  # Invert y-axis to have the highest accuracy at the top

# Add labels to the bars
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, round(acc, 2), 
             va='center', ha='left', fontsize=8, color='black')

plt.show()
```
